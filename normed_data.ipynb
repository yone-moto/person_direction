```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv("./input/pose_angle_test_ffill.csv")
df_train
```

```python
x_list = []
y_list = []
for j in range(25):
    x_list.append(f"x_{j}")
for j in range(25):
    y_list.append(f"y_{j}")
df_train_x = df_train[x_list]
df_train_y = df_train[y_list]
df_train_x
```

```python
np_train_x = np.array(df_train_x)
for i,train_x in enumerate(np_train_x):
    df_train_x[i:i+1] = (df_train_x[i:i+1] - np.nanmin(train_x))/(np.nanmax(train_x) - np.nanmin(train_x))
```

```python
df_train_x
```

```python
np_train_y = np.array(df_train_y)
for i,train_y in enumerate(np_train_y):
    df_train_y[i:i+1] = (df_train_y[i:i+1] - np.nanmin(train_y))/(np.nanmax(train_y) - np.nanmin(train_y))
```

```python
df_tmp = pd.concat([df_train_x,df_train_y],axis=1)
df_tmp["angle"] = df_train["angle"]
```

```python
df_tmp["angle"].unique()
```

```python
df_inv = df_tmp.copy()
flip_vec = (0, 1, 5, 6, 7, 2, 3, 4, 8, 12, 13, 14, 9, 10, 11,16, 15, 18, 17, 22, 23, 24, 19, 20, 21)
for i,inv in enumerate(flip_vec):
    df_inv[f"x_{inv}"] = df_tmp[f"x_{i}"]
    df_inv[f"x_{inv}"] = df_tmp[f"x_{i}"]
df_inv["angle"] = -df_tmp["angle"]
df_inv
```

```python
df_inv = df_inv[df_inv["angle"] != 0.0]
df_inv = df_inv[df_inv["angle"] != 180.0]
df_inv = df_inv[df_inv["angle"] != 90.0]
df_inv = df_inv[df_inv["angle"] != -90.0]


df_inv["angle"].unique()
```

```python
df_final = pd.concat([df_inv,df_tmp]) 
df_final["angle"].unique()
```

```python
df_final.groupby("angle")["angle"].count()
```

```python
df_final["angle"] = df_final["angle"].replace(0,0).replace(30,1).replace(60,2).replace(90,3).replace(120,4).replace(150,5).replace(-180,6).replace(-150,7).replace(-120,8).replace(-90,10).replace(-60,11).replace(-30,12)
```

```python
# df_final = df_tmp.fillna(-1)
df_final.to_csv("./input/df_test_final.csv",index=False)
```
